### 2
Выберите верные утверждения о дисперсии коэффициента регрессии $\hat\beta_j$, получаемого методом наименьших квадратов.

$$ D(\hat\beta_j) = \frac{\sigma^2}{TSS_j(1 - R^2_j)}$$
$$ TSS = \sum\limits_i^n(y_i - \hat{y})^2 $$
$$ R^2 = \frac{ESS}{TSS} = 1 - \frac{RSS}{TSS} $$


- Чем больше $R_j$, коэффициент детерминации при регрессии $x_j$ на все остальные признаки, тем больше дисперсия $\hat{\beta}_j$ $\large\checkmark$


- Чем сильнее признак $x_j$ варьируется в выборке, тем меньше дисперсия $\hat{\beta}_j$   $\large\checkmark$


- Чем больше объём выборки $n$, тем меньше дисперсия $\hat{\beta}_j$   $\large\checkmark$


- Чем больше дисперсия ошибки $\varepsilon$, тем больше дисперсия $\hat{\beta}_j$  $\large\checkmark$
