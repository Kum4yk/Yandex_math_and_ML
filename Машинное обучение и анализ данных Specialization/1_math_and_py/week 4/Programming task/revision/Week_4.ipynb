# Экспоненциальное распределение

Экспоненциальное (или показательное) распределение — абсолютно непрерывное распределение, моделирующее время между двумя последовательными свершениями одного и того же события

Построим график функции распределения:

```python
x = np.linspace(0,4,100)
cdf = sts.expon.cdf(x) 
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

Сгенерируем из него выборку объёма 1000, построим гистограмму выборки и нарисуем поверх неё теоретическую плотность распределения случайной величины

```python
 data = sts.expon.rvs(size=1000)
```

```python
x = np.linspace(0,4,100)
pdf = sts.expon.pdf(x)
plt.plot(x, pdf)
plt.hist(data, density=True)
plt.ylabel('fraction of samples, pdf')
plt.xlabel('$x$')
```

При пяти значениях n (5, 10, 50, 100, 1000) сгенерируем 1000 выборок объёма n и построим гистограммы распределений их выборочных средних.

Среднее значение  экспонициалиного распределения равно:   1/Lamda
    
Дисперсия  экспонициалиного распределения равна:   1/Lamda**2

При генерации выборок я задал lambda=1

Тогда средние значения выборок будет стремится к нормвльному распределению с параметрами:

m=1, sigma=1/sqrt(n). Я добавил графики плотности нормального распределения с указанными параметрами к гистограммам  распределений их выборочных средних для всех значений n.

```python
n = 5
average =[]
for i in range(1000):
    data = sts.expon.rvs(size=n)
    average.append(statistics.mean(data))
plt.hist(average,density=True)
norm = sts.norm(1, 1/n)
pdf = norm.pdf(x[:300])
plt.plot(x[:300], pdf)
plt.ylabel('$F(x)$    N='+str(n)+',    $Norm$')
plt.xlabel('$x$')
```
