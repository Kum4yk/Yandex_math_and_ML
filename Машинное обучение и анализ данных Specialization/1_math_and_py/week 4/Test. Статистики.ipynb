10. Предположим, что вы получаете в среднем три спам-письма в день. Какова доля дней, в которые вы получаете пять или больше спам-писем? Округлите до двух знаков после десятичной точки.
___

Вероятность 5 или больше  = 1 - сумма вероятностей от 0  до 4


```python
def pois(numbers, mean=1):
    import math
    return math.pow(mean, numbers) * math.exp(-mean) / math.factorial(numbers)


summ = 0.0
for i in range(5):
    summ += pois(i, mean=3)
print(round(1 - summ, 2))
```
