**1. Начальные приближения центров кластеров в K-Means лучше брать:**


- Достаточно удаленными друг от друга. $\large \checkmark $


- Близкими друг к другу.


- Подальше от основной массы объектов из выборки.


**2. В K-Means++ вероятность выбрать точку выборки как центр кластера пропорциональна:**


- Расстоянию от точки до ее ближайшего соседа среди всех точек выборки.


- Расстоянию от точки до самой удаленной от нее точки выборки.


- Натуральному логарифму расстояния от точки до K-того соседа.


- Квадрату расстояния от точки до ближайшего к ней центра кластера. $\large \checkmark $


**3. В агломеративной иерархической кластеризации:**


-  На первой итерации есть только один кластер.


- На первой итерации количество кластеров совпадает с количеством точек в выборке. $\large \checkmark $


- На каждой итерации объединяются два ближайших друг к другу кластера. $\large \checkmark $


- На каждой итерации объединяются несколько (возможно, более двух) ближайших друг к другу кластера.



**4. В кластеризации с помощью EM-алгоритма решается задача разделения смеси распределений. В качестве компонент смеси выступают:**


- Плотности распределения отдельных признаков.


- Плотности распределения вектора признаков в кластере. $\large \checkmark $



- Апостериорные вероятности принадлежности объектов выборки к кластеру.



**5. Если выбирать количество кластеров по среднему внутрикластерному расстоянию, наилучшим числом кластеров будет:**


- 1


- 2


- Половина от числа точек (с округлением вверх для нечетного числа)


- Количество точек в выборке. $\large \checkmark $
