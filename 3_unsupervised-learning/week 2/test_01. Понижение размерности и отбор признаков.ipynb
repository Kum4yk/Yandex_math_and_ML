- Жадные методы всегда находят то подмножество признаков, которое является лучшим для данного метода обучения.

**6. Выберите верные утверждения про метод главных компонент:**


- Метод главных компонент — это метод понижения размерности с потерями, если число новых признаков d меньше ранга исходной матрицы "объекты-признаки".  $\large \checkmark$


- Метод главных компонент является линейным методом понижения размерности — новые признаки являются линейными комбинациями исходных признаков.  $\large \checkmark$


- Метод главных компонент позволяет найти нелинейную поверхность, при проецировании выборки на которую дисперсия оказывается максимальной.  


- Матрица весов $W$, которая строится в методе главных компонент, является ортогональной.  $\large \checkmark$
