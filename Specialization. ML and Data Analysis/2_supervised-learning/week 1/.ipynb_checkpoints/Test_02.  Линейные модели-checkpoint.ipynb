1. В чём заключается отличие градиентного спуска от стохастического градиентного спуска?
 - В стохастическом градиентном спуске на каждой итерации делается шаг в случайном направлении.
 - В стохастическом градиентном спуске на каждой итерации используется лишь одно слагаемое в функционале ошибки. + 
 - В стохастическом градиентном спуске на каждой итерации к антиградиенту добавляется нормальный шум.



2. Как можно избавиться от свободного члена в линейных моделях?
 - Добавить единичный признак. + 
 - Свободный член нельзя убрать, не ослабив при этом линейную модель.
 - Просто убрать его — свободный член ни на что не влияет и нужен лишь для простоты выкладок.



3. Выберите верные утверждения про отступ
 - Отступ характеризует качество классификатора на всей обучающей выборке. (-)
 - Если отступ на объекте положительный, то классификатор даёт верный ответ на данном объекте. (+) ()
 - Чем больше абсолютное значение отступа, тем больше классификатор уверен в своём ответе. (+)
 - Отступ характеризует качество классификатора на одном объекте. (+) 

4. Для чего при обучении линейных классификаторов используются верхние оценки на пороговую функцию потерь?
 - Чтобы заменить разрывную пороговую функцию потерь на гладкую функцию — без этого градиентная оптимизация будет невозможна. (+)
 - Чтобы перейти к функции потерь, которая будет легко вычислима — без этого градиентная оптимизация будет занимать слишком много времени. 
 - Чтобы перейти к функции потерь, на которой градиентный спуск будет быстрее сходиться.
