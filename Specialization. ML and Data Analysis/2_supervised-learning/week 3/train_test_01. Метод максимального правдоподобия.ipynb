**1. В сумке Кати $N$ похожих ключей; только один из них подходит к двери от дома. Когда Катя вечером приходит домой, она наугад достаёт ключ из сумки и пытается открыть им дверь. Если ключ не подходит, она кладёт его обратно в сумку и снова достаёт наугад следующий. Каким законом задаётся распределение номера попытки $X$, с которой Кате удаётся открыть дверь?**  


- $ P\left(X=k\right)=\frac1{N}\left(1-\frac1{N}\right)^{k-1}$  
Верно. Такое распределение, кстати, называется геометрическим.
 

**2. Представим, что у нас есть данные $ X_1,\dots,X_n $ за $n$ дней о том, с какой попытки Кате удалось открыть дверь. Запишите функцию правдоподобия такой выборки и возьмите от неё логарифм. Какое выражение у вас получилось?**


- $ \ln L\left(X^n, N\right) = n \ln \frac1{N} + \left(\sum\limits_{i=1}^n X_i - n\right) \ln \left(1-\frac1{N}\right)$



$$ \ln L\left(X^n, N\right) = \prod\limits_{i=1}^n \frac{1}{N}\left(1 - \frac{1}{N}\right)^{X_i - 1} = \frac{1}{N}^n \left( 1 - \frac{1}{N} \right)^{\sum\limits_{i=1}^{n}X_i - n} $$

**3. За последние пять дней Кате удавалось открыть дверь с восьмой, двенадцатой, седьмой, шестой и снова двенадцатой попыток; давайте найдём оценку максимального правдоподобия для N. Подставьте эти числа в выражение для логарифма правдоподобия, продифференцируйте его по N и приравняйте к нулю. Из полученного уравнения выразите N. Итак, $\hat{N}_{ОМП}$ = :**
$$ 9 $$


Подставляем числа в выражение для логарифма правдоподобия:
$$ \ln L\left(X^n, N\right) = 5 \ln \frac1{N} + 40\ln \left(1-\frac1{N}\right) $$

берём производную по N, приравниваем к нулю и получаем:

$$ 40N = 5N(N-1)$$
$$N=9$$

Кстати, и в этом случае оценка максимального правдоподобия совпала с выборочным средним.
