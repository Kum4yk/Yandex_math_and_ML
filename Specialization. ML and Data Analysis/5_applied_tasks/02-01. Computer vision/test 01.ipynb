### 7.
Для повышения контраста можно использовать разнообразные функции преобразования яркости. Такие функции g(x) должны удовлетворять единственному условию - их выход должен лежать в диапазоне [0, 255], для любого x из диапазона [0,255], при этом g(0) = 0, g(255) = 255.

Если в качестве такой функции выбрать семейство $с*log(1 + a*x)$, где x - яркость входного изображения, а a и c константы. a, c > 0.
    
- увеличивают в темных областях и уменьшают в светлых
