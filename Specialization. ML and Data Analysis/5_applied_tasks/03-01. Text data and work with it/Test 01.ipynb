### 1
Чем n-граммы отличаются от skip-грамм?


- k-skip-n-граммы — это такие наборы из n-грамм, что расстояние между ними не превышает k


- k-skip-n-граммы — это такие наборы из k слов, что расстояние между соседними словами из набора не превышает n


- k-skip-n-граммы — это такие наборы из n слов, что расстояние между соседними словами из набора не превышает k $\large \checkmark$


### 2
В чём заключаются преимущества хэширования слов?


- Хэширование позволяет сгруппировать слова, которые похожи по смыслу


- При хэшировании не нужно хранить соответствие между токенами и номерами признаков $\large \checkmark$


- При хэшировании можно получить более компактное представление объектов без потери информации


- Хэширование позволяет уменьшить количество признаков $\large \checkmark$

### 3
Какие проблемы могут возникнуть при применении деревьев к текстовым данным?


- Низкая скорость обучения из-за большого количества признаков $\large \checkmark$


- Необходимость построения большого количества деревьев при обучении композиций $\large \checkmark$


- Необходимость построения деревьев большой глубины для учёта взаимодействий между признаками $\large \checkmark$
