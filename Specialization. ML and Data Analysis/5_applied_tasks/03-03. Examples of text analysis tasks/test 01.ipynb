Почему в аннотировании текстов часто применяют unsupervised методы? (выберите наиболее точный ответ)


- Unsupervised методы как правило лучше работают, чем supervised


- Unsupervised методы в задаче аннотирования будут работать лучше, чем supervised, даже обученные на большой обучающей выборке


- Unsupervised методы в условиях наличия хорошей обучающей выборки работают хуже, а в условиях ее маленького размера - лучше supervised. Выборку в этой задаче собирать сложно, поэтому как правило и прибегают к unsupervised методам. +
