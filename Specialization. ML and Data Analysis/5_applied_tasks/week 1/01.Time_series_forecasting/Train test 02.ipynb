**9**

Сложный вопрос: какое начальное приближение для параметра $p$ подсказывает нам частичная автокорреляционная функция?

![](../../Data/1_7.png)

У нашего ряда много несезонных лагов, достаточно сильно отличающихся от нуля — это следствие наличия в ряде циклов. Методология выбора гиперпараметров, тем не менее, остаётся прежней.

- 1 - 


- 11 +


- 12


- 13 (Этот лаг действительно значим, но он слишком большой! Если мы выберем такое $p$, то модель будет неоднозначно определённой: $y_{12}$ будет входить и в сезонную, и в несезонную компоненты, поэтому коэффициенты невозможно будет определить. По этой причине при $P>0$ берут $p<S$.)
