#### 4. Укажите верные утверждения:


- В MDS расстояние между объектами в исходном пространстве измеряется по евклидовой метрике.


- В MDS расстояние между объектами в новом (маломерном) пространстве измеряется по евклидовой метрике. **+**


- MDS старается так отобразить объекты в маломерное пространство, чтобы как можно точнее сохранить расстояния между объектами в исходном пространстве. **+**


- MDS старается так отобразить объекты в маломерное пространство, чтобы как можно точнее сохранить пропорции в расстояниях между объектами.

#### 7. Методу t-SNE, в отличие от SNE, удается сохранить промежутки между кластерами в маломерном пространстве. За счет чего это достигается?


- Функционал в t-SNE меньше штрафует увеличение расстояний между объектами в маломерном пространстве, поэтому этот метод "не боится" располагать точки далеко друг от друга
