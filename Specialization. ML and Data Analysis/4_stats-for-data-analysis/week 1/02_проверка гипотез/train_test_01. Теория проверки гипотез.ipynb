### 1
Если $p\leq\alpha $ то:

 - $H_0$ отвергается в пользу $H_1$
---
---

### 2
Достигаемый уровень значимости — это:

 - ничего из перечисленного
 ---
 ---

### 3
Ошибка первого рода — это:


- отвержение верной нулевой гипотезы
---
---

### 4
Выберите верные утверждения про пару $(H_0, H_1)$

- Порядок в паре $(H_0, H_1)$ выбирается так, чтобы $H_0 $ значила "ничего не происходит", а $H_1$ "мы сделали важное открытие", потому что с помощью проверки гипотез нельзя доказать справедливость $H_0$: при $p \leq \alpha$, скорее всего, верна $H_1$ , а при $p>\alpha$ либо верна $H_0$, либо данных недостаточно.


- $H_0$ в каком-то смысле для нас важнее: мы верим в неё по умолчанию, и отказываемся от неё только если есть хорошие доказательства 

---
---

### 5
Что такое мощность статистического критерия?

- Вероятность отвергнуть неверную нулевую гипотезу

---
---

### 6
Какое свойство статистического критерия важнее?


- Корректность: вероятность ошибки первого рода не должна превышать $\alpha$ 
---
---

### 7
Выберите верные утверждения про статистическую и практическую значимость.


- Размер эффекта нас интересует в первую очередь, а достигаемый уровень значимости — это просто способ удостовериться, что мы его достаточно точно оценили


- С ростом объёма выборки эффекты всё более незначительных размеров будут признаваться статистически значимыми
