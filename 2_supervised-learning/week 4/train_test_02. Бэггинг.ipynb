2. **Что характеризует компонента разброса в разложении ошибки?**


 - Дисперсию прогнозов алгоритмов, обученных по разным выборкам — чем сильнее изменяются прогнозы при изменении обучающей выборки, тем больше разброс. **(+)**


 - Дисперсию прогнозов алгоритмов, обученных по разным подмножествам признаков — чем сильнее изменяются прогнозы при исключении тех или иных признаков, тем больше разброс.


 - Дисперсию прогнозов алгоритма на обучающей выборке — чем больше отклонения прогнозов от среднего, тем больше разброс.

___

3. **Что характеризует компонента смещения в разложении ошибки?**


- Отклонение прогнозов алгоритмов, обученных по разным выборкам, от прогнозов оптимального алгоритма — чем лучше эти алгоритмы в среднем приближают оптимальный, тем ниже смещение. **(+)**


 - Отклонение прогнозов алгоритмов, обученных по разным подмножествам признаков, от прогнозов оптимального алгоритма — чем лучше эти алгоритмы в среднем приближают оптимальный, тем ниже смещение


 - Отклонение прогнозов нашего алгоритма на обучающей выборке от прогнозов оптимального алгоритма — чем меньше отклонение, тем ниже смещение.
