1. **Чем градиентный бустинг отличается от случайного леса?**


 - Градиентный бустинг может строить алгоритмы только для задач регрессии.


 - Каждый следующий алгоритм в градиентном бустинге обучается так, чтобы исправить ошибки предыдущих базовых алгоритмов. **$\checkmark$**


 - Базовые алгоритмы, как правило, выбираются достаточно простыми — например, это могут быть неглубокие деревья. **$\checkmark$**

___

2. **Градиент какой функции вычисляется на каждой итерации градиентного бустинга?** 


 - Функции, которая вычисляет ошибку композиции на обучающей выборке. **$\checkmark$**


 - Функции, которая вычисляет прогноз композиции в зависимости от признакового описания объекта.


 - Функции, которая вычисляет ошибку композиции на одном из объектов обучающей выборки.

___

3. **Для чего нужно сокращение шага в градиентном бустинге?**


 - Понижение вклада каждого базового алгоритма в общий прогноз. **$\checkmark$**


 - Учёт особенностей функции потерь при построении композиции. - 


 - Борьба с переобучением. **$\checkmark$**


 - Ускорение построения композиции.

___

4. **В чём заключается переподбор прогнозов в листьях дерева в градиентном бустинге?**


 - В каждом листе выбирается прогноз, оптимальный с точки зрения исходной функции потерь, а не с точки зрения среднеквадратичной ошибки. **$\checkmark$**


 - В каждом листе выбирается прогноз, оптимальный с точки зрения среднеквадратичной ошибки, тогда как по умолчанию этот прогноз просто равен среднему ответу на объектах в листе.


 - В каждом листе выбирается прогноз, оптимальный с точки зрения регуляризованной функции ошибки.


 - В каждом листе прогноз домножается на некоторую константу между нулём и единицей, за счёт чего понижается риск переобучения.
