1. Вам нужно предсказать, повысится или понизится завтра курс доллара. Какая это задача?
 - Регрессия
 - Классификация + 
 - Кластеризация
___
В задаче всего два ответа (курс повысится и курс понизится), поэтому она является задачей классификацией.

2. Вам нужно предсказать, каким завтра будет курс доллара. Какая это задача?
 - Регрессия +
 - Классификация
 - Кластеризация
___
Курс доллара принимает вещественные значения, поэтому мы имеем дело с задачей регрессии.

 - Кластеризация +
