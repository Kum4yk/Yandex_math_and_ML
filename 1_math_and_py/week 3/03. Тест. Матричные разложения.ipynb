
#### Вопрос 1
#####  Произведением каких матриц представляется исходная при сингулярном разложении (SVD)? 

- Диагональной и ортогональной.
- Ортогональной, диагональной и ортогональной.
- Диагональной, ортогональной и диагональной. (+, Правильно. При этом если взять из них подматрицы и перемножить, то можно получить хорошее приближение исходной матрицы матрицей меньшего ранга.)
- Нижней треугольной и верхней треугольной.

#### Вопрос 2
##### Чем позволяет приблизить исходную матрицу сингулярное разложение?
- Матрицей меньшего ранга. (+, Верно! При этом такая матрица является оптимальной среди всех матриц с таким рангом по норме Фробениуса)
- Матрицей меньшего размера.
- Ортогональной матрицей.

#### Вопрос 3
##### Как с матричными разложениями связана задача рекомендации фильмов пользователям?
- Выполняется разложение исходной матрицы пользователи-фильмы (с оценками фильмов) на произведение матрицы профилей пользователей и матрицы профилей фильмов. Произведение профиля пользователя на профиль фильма — это прогноз оценки, которую пользователь поставит фильму. Рекомендуются фильмы, для которых этот прогноз больше, либо этот прогноз можно подавать на вход другому алгоритму. (+)
- Выполняется сингулярное разложение матрицы с оценками фильмов. Из сингулярного разложения берется диагональная матрица, значения в ней - рейтинги фильмов. Рекомендуется фильм, у которого больше рейтинг, либо можно подавать эти рейтинги на вход другому алгоритму.
