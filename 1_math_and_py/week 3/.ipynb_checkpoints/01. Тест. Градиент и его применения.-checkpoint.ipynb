### Градиент и его применения
##### Вопрос 1. Градиент — это
 - вектор нормали к поверхности, которая является графиком функции
 - один из направляющих векторов касательной плоскости к графику функции.
 - частная производная функции.
 - производная функции по заданному направлению. 
 - вектор, координатами которого являются частные производные функции по разным аргументам. (+)

##### Вопрос 2. Геометрический смысл градиента заключается в том, что
 - градиент задает направление наискорейшего убывания функции.
 - градиент задает направление наискорейшего роста функции. (+)
 - градиент задает направление линий уровня функций.

##### Вопрос 3. Поиск минимума функции методом градиентного спуска заключается в итеративном движении в направлении 
 - градиента функции. (-) т.к. это направление максимума, если он существет
 - антиградиента (т.е. минус градиента) функции. (+)

##### Вопрос 4. Как себя должен вести шаг градиентного спуска? Выберите все верные замечания.
 - Сначала быть небольшим, затем увеличиваться. (-)
 - Уменьшаться по мере приближения к экстремуму, либо быть постоянным, но небольшим. (+) повышаем точность, опасность проскочить
 - Быть положительным. (-)
 - Принимать знак в зависимости от значения и направления градиента. (+) звучит логично, т.к. речь про f(x + $\delta$x)
