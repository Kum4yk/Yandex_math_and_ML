#### 1. Как определяется ответ решающего дерева в задаче классификации для заданного объекта?


 - Ответ определяется на основе суммы значений из всех вершин, по которым прошел данный объект


 - По значению в листе, в который попал данный объект.   
 (+ Объект опускается из корневой вершины в один из листьев согласно правилам, записанным в вершинах дерева. Ответ определяется на основе значения в этом листе.)

    
 - Ответ определяется на основе значения в любой из вершин дерева

#### 2. Какие из нижеперечисленных пунктов описывают достоинства решающих деревьев?


 - Устойчивость к масштабу признаков   
 (+ Поскольку решающие деревья состоят из логических правил вида "значение признака меньше порога", они устойчивы к изменению масштаба признаков. Это их преимущество — не нужно беспокоиться о предобработке данных.)


 - Возможность работы с данными с пропусками   
 (+ Деревья действительно могут работать с пропущенными значениями признаков.)


 - Чувствительность к составу выборки (-, Недостаток, правда)


 - Высокий риск переобучения (-, Недостаток, правда)


 - Интерпретируемость модели   
 (+ Деревья представляют собой набор простых логических правил. Решение, принятое деревом, можно объяснить пользователю благодаря этому.)


 - Хорошо работают на выборках с большим количеством признаков (-)
