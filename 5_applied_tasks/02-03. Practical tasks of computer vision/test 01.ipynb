### 2

Часто в качестве последнего слоя в задаче классификации используется слой softmax который выглядит как:
$$ h_j = \frac{e^{z_j}}{\sum e^{z_i}}$$
где $z_i$ входы этого слоя, а $h_j$ выходы. Номер, выхода с максимальным значением соответствует номеру предсказанного класса.

- выходы softmax слоя всегда $>0$ (по идее может быть нулевой) $\leq1$ $\large \checkmark$
  
  
- выходы softmax слоя всегда $\leq1$ $\large \checkmark$


- softmax монотонная функция, то есть выход с максимальным значением соответствует максимальному значению входа $\large \checkmark$


- сумма выходов слоя равна 1 $\large \checkmark$
